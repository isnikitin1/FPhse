```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.zeros((15, 28))
a[2:-2,1]=1; a[2,2:6]=1
a[2:7,6]=1; a[7:-2,7]=1
a[7,2:7]=1; a[-3,2:7]=1
a[2:-2,10]=1; a[2:-2,14]=1 
a[2:-2,18]=1; a[-3,10:19]=1
a[2,21:-2]=1; a[-3,21:-2]=1
a[7,21:-2]=1; a[2:-2,-2]=1
plt.imshow(a)
```

```python
u, s, vh = np.linalg.svd(a)
rank = np.linalg.matrix_rank(a)
print("rank(a) =", rank)
```

```python
vt = vh.T
anew = 0
for i in range(rank):
    vtn = vt[:, i]
    un = u[:, i]
    anew = anew + s[i] * (un[:, np.newaxis] @ vtn[np.newaxis, :])
    plt.imshow(anew)
    plt.show()
```
