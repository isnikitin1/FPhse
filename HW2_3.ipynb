```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.patches as mt

a = np.array([[1, 1], [1, 1]])

u, s, vh = np.linalg.svd(a)
v = vh.T.conj()
fig = plt.figure(figsize=(7,7))
ill = plt.subplot(aspect='equal')
ellipse = mt.Ellipse((0, 0),2 ,2 ,facecolor = 'none',edgecolor = 'green') 
ill.add_artist(ellipse)
plt.arrow(0, 0, *(v[:, 0]), color = 'red')
plt.arrow(0, 0, *(v[:, 1]), color = 'blue')
plt.ylim(-1.5,1.5)
plt.xlim(-1.5,1.5)
plt.grid()
u
```

```python
fig = plt.figure(figsize=(7,7))
ill = plt.subplot(aspect='equal')
x = np.array([1,0])
ellipse = mt.Ellipse((0, 0),2 * s[1],2 * s[0], angle = 180/math.pi*math.acos(u[:, 1]@x),facecolor = 'none',edgecolor = 'green')
ill.add_artist(ellipse)
plt.arrow(0, 0, *s[0]*u[:, 0], color = 'red')
plt.arrow(0, 0, *s[1]*u[:, 1], color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(-1.5, 1.5) 
plt.grid()
```
